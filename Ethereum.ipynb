```python
import seaborn as sns
import matplotlib.pyplot as plt

# Obtener la lista de columnas excluyendo la primera
columns = df.columns[1:]

# Eliminar duplicados
df.drop_duplicates(inplace=True)

# Calcular el número de filas y columnas para la disposición de subgráficas
nrows = (len(columns) + 2) // 3
ncols = 3

# Crear la figura y los ejes
fig, axes = plt.subplots(nrows, ncols, figsize=(16, 16), constrained_layout=True)
plt.subplots_adjust(wspace=0.7, hspace=0.8)
plt.suptitle("Distribución de los atributos(Escala logarítmica)", y=0.95, size=18, weight='bold')

# Bucle para crear boxplots para cada columna
for i, col_name in enumerate(columns):
    row = i // ncols
    col = i % ncols
    ax = axes[row, col]
    sns.boxplot(ax=ax, data=df, x=col_name)
    ax.set_title(f'Distribución de {col_name}')
    ax.set_xscale("log") # Opcional: Establece la escala logarítmica en el eje x

# Eliminar los subgráficos vacíos si hay alguno
for j in range(i+1, nrows*ncols):
    fig.delaxes(axes[j // ncols, j % ncols])

# Mostrar la figura
plt.show()
```

```python
from sklearn.model_selection import ParameterGrid
import xgboost as xgb
# Define the parameters for the search
parameters = {
    'max_depth': [3,6,9],
    'learning_rate': [0.12,0.2],
    'subsample': [0.2,0.5,0.8],
    'n_estimators': [100,300,600],
   
}

results = []

# Create a list of all possible combinations of parameters
parameter_combinations = list(ParameterGrid(parameters))

# Iterate over each parameter combination
for params in parameter_combinations:

    # Set up the XGB classifier with the current parameters
    xgb_clf = xgb.XGBClassifier(eval_metric="auc",objective= "binary:logistic", **params)

    # Train the XGB classifier
    xgb_clf.fit(x_tr_resample, y_tr_resample)

    # Predict with the model
    preds = xgb_clf.predict(X_test)

    # Get the classification report
    report = classification_report(y_test, preds, output_dict=True)

    # Flatten the classification report
    flat_report = {}
    for k, v in report.items():
        if isinstance(v, dict):
            for subk, subv in v.items():
                flat_report[f"{k}_{subk}"] = subv
        else:
            flat_report[k] = v

    # Save the results
    result = {
        'parameters': params,
        **flat_report,
    }
    results.append(result)

    # Print the results
    print(f"Parameters: {params}, Accuracy: {result['accuracy']}, Macro Avg Precision: {result['macro avg_precision']}, Macro Avg Recall: {result['macro avg_recall']}, Macro Avg F1-score: {result['macro avg_f1-score']}", report)

# Convert the results to a DataFrame
df_results = pd.DataFrame(results)


```
